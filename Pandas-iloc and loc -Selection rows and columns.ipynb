```python
import pandas as p
df=p.read_csv("x.csv")
df.iloc[:]       # iloc[rows slice,columns Slice]   iloc[:]->select all rows and columns
                 # numerical value pass only in iloc 
                 # basically used for row wise indexing selection
```

```python
import pandas as p
df=p.read_csv("x.csv")
df.iloc[3:4,1]="Vai"              # find particular index and then change value by assign method
df
df.iloc[3,1]="By"                 # simply way to get particular index
df
```

```python
import pandas as p
df=p.read_csv("x.csv")
df.iloc[3,0:2]=["Vai",'Ban'] # multiple values change
df
```

```python
import pandas as p
df=p.read_csv("x.csv")
df.loc[:,"Name"]       # loc[rows slice,Columns Slice] - use to select data
                       # only alphabets indexing can pass
                       # basically we can use it for columns slicing
```
