```python
import pandas as p
df=p.read_csv("new_x.csv")
y=p.crosstab(df.Age,df.Passout)
             # pandas.crosstab(dataframe.column_name(Row),dataframe.column_name(column))
print(y)
            # age crosscheck with passout column, two way crosstab
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
y=p.crosstab([df.Age,df.Passout],df.Name)
         #pandas.crosstab([list of columns],columns)
y        
         #three way crosstab
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
y=p.crosstab(df.Age,df.Passout,margins=True) 
            # here margins=True help to get total values -All shows it 
y           #two way crosstab
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
y=p.crosstab([df.Age,df.Passout],df.Name,margins=True) 
            # here margins=True help to get total values -All shows it 
y           #three-way crosstab
```
