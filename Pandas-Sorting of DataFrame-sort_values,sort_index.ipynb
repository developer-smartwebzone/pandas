```python
import pandas as p
df=p.read_csv("new_x.csv")
df.sort_values(by='Age')    # .sort_values(by="Column_name")
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
df.sort_values(by='Age', ascending=False)    # .sort_values(by="Column_name",ascending =False)
                                             # use to get values in decending order
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
df.index=[2,3,5,6,1,0,4]   # change index 
print(df)
df.sort_index()     # .sort_index()-> sort according to index
```

```python
import pandas as p
df=p.read_csv("new_x.csv")
df.index=[2,3,5,6,1,0,4]   # change index 
print(df)
df.sort_index(ascending=False)     # .sort_index(ascending=False)-> sort according in decending order
```
